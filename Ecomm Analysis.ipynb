```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns
```

```python
dataset = pd.read_excel("ECOM_DATA.xlsx")
```

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

```python
dataset = dataset.drop('Postal Code', axis=1)
```

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

# Order Priority

```python
dataset['Order Priority'].value_counts()
```

```python
plt.figure(figsize = (4,3))
sns.countplot(x="Order Priority", data = dataset)
plt.title("COUNT OF ORDER PRIORITY")
plt.show()
```

# Ship Mode

```python
dataset['Ship Mode'].value_counts()
```

```python
x = dataset['Ship Mode'].value_counts().index
y = dataset['Ship Mode'].value_counts().values
```

```python
plt.figure(figsize = (5,4))
plt.pie(y,labels=x,startangle = 60,autopct="%0.2f%%")
plt.legend(loc=2)
plt.show()
```

```python
plt.figure(figsize = (5,4))
sns.countplot(x="Ship Mode",data=dataset,hue="Category")
plt.title("COUNT OF SHIPMODE BY CATEGORY")
plt.show
```

# customer segment

```python
plt.figure(figsize = (6,4))
sns.countplot(x="Segment",data=dataset  )
plt.title("COUNT OF SEGEMENT")
plt.show
```

# product category

```python
plt.figure(figsize = (6,4))
sns.countplot(x="Category",data=dataset  )
plt.show
```

```python
plt.figure(figsize = (10,6))
sns.countplot(x="Category",data=dataset[dataset["Category"]=="Office Supplies"],hue="Sub-Category"  )
plt.title("COUNT OF OFFICE SUPPLIES BY CATEGORY")
plt.show
```

```python
dataset["Order year"] = dataset['Order Date'].dt.year
```

```python
dataset['Order year'].value_counts()
```

```python
plt.figure(figsize = (6,4))
sns.countplot(x="Order year",data=dataset  )
plt.show
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x="Category",y="Profit",data=dataset,estimator="sum")
plt.title("PROFIT BY CATEGORY")
plt.show
```
